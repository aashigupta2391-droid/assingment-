1. what is difference between github and git?
- Git is a distributed version control system that tracks changes in source code during software development. It allows multiple developers to work on the same project simultaneously without overwriting each other's changes. GitHub, on the other hand, is a web-based hosting service for Git repositories. It provides a graphical interface, collaboration features like pull requests and issue tracking, and allows developers to share their Git projects with others.

- A pull request (often abbreviated as PR) is a feature in GitHub (and other Git hosting services like GitLab and Bitbucket) that allows developers to propose changes to a repository's codebase. Here's a breakdown of the concept:

Forking and Branching: Typically, a developer first 'forks' the main repository, creating a personal copy. Then, they create a new 'branch' within their forked repository to isolate their changes from the main codebase.
Making Changes: The developer then makes their desired code changes, bug fixes, or new features on this dedicated branch.
Opening a Pull Request: Once the changes are complete and tested, the developer opens a pull request. This is essentially a formal request to the maintainers of the original repository to 'pull' their changes from their branch into the main codebase.
